```python
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.api import VARMAX
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# Load data
dataset = pd.read_csv("Tatacoffee13_21.csv", parse_dates=['Date'], index_col='Date')

# Scale data
Ms = MinMaxScaler()
data = Ms.fit_transform(dataset)
data = pd.DataFrame(data, columns=["Open", "High", "Low", "Close"])

# Add exogenous variables (example: dummy exogenous data)
exog_data = np.random.randn(len(data), 2)  # Example: 2 exogenous variables
exog = pd.DataFrame(exog_data, columns=["Exog1", "Exog2"], index=data.index)

# Initialize performance dictionary
performance = {"Model": [], "RMSE": [], "MaPe": [], "Lag": [], "Test": []}

# Select columns for modeling
listt = ["Close", "High", "Open", "Low"]
datasetTwo = data[listt]

# Train-test split
test_obs = 100
train = datasetTwo[:-test_obs]
test = datasetTwo[-test_obs:]
train_exog = exog[:-test_obs]
test_exog = exog[-test_obs:]

# VARMAX model
order = (1, 1)  # AR order = 1, MA order = 1 (you can tune this)
model = VARMAX(train, exog=train_exog, order=order)
result = model.fit(disp=False)  # Disable verbose output

# Forecasting
pred = result.forecast(steps=test_obs, exog=test_exog)
preds = pd.DataFrame(pred, columns=listt)
preds.to_csv("varmax_forecasted_{}.csv".format(test_obs))

# Performance evaluation
rmse = round(mean_squared_error(test, pred, squared=False))
mape = mean_absolute_percentage_error(test, pred)

# Store performance metrics
performance["Model"].append("VARMAX")
performance["RMSE"].append(rmse)
performance["MaPe"].append(mape)
performance["Lag"].append(order)
performance["Test"].append(test_obs)
perf = pd.DataFrame(performance)

print(perf)
```
